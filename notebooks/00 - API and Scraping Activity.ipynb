<h1>Pokemon Generations</h1>


```python
import requests
from bs4 import BeautifulSoup

URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')

tables = soup.find(id='mw-content-text').find_all('table')
len(tables)
# 1 - 8 are tables with pokemon in each generation
```

```python
tables[1].contents[7].contents
```

```python
tables[1].contents[7].contents[7].text.strip()
```

```python
pokemon_tables = tables[1:9]
len(pokemon_tables)
```

```python
pokemon_json = []

for table in pokemon_tables:
    for x in range(3, len(table), 2):
        kdex=table.contents[x].contents[1].text.strip()
        ndex=table.contents[x].contents[3].text.strip()
        name=table.contents[x].contents[7].text.strip()
        type1=table.contents[x].contents[9].text.strip()
        if len(table.contents[x]) > 10:
            type2=table.contents[x].contents[11].text.strip()
            pokemon_json.append({
                "kdex": kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "type2": type2
            })
        else:
            pokemon_json.append({
                "kdex": kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1
            })

pokemon_json
```

```python
import json

with open('pokemon_json.json', 'w') as outfile:
    json.dump(pokemon_json, outfile)
```

<h1> News </h1>

We will we getting new details via NewsAPI.

```python
import requests, json

responses11 = json.loads(requests.get("https://newsapi.org/v2/everything?sources=techcrunch&from=2021-03-11&to=2021-03-11&sortBy=publishedAt&apiKey=ab5d03d7bc6e4e7d8c3678703f388e41").text)

responses12 = json.loads(requests.get("https://newsapi.org/v2/everything?sources=techcrunch&from=2021-03-12&to=2021-03-12&sortBy=publishedAt&apiKey=ab5d03d7bc6e4e7d8c3678703f388e41").text)
```

```python
responses = []
responses.append(responses11)
responses.append(responses12)
responses
```

```python
# clean dataset to only get date, title, full article, author
to_json = []
news_object = {}

for table in responses:
    for news in table["articles"]:
        news_object["date"] = news["publishedAt"]
        news_object["title"] = news["title"]
        news_object["full_article"] = news["content"]
        news_object["author"] = news["author"]
        to_json.append(news_object)
to_json
```

```python
with open('techcrunch_news.json', 'w') as outfile:
    json.dump(responses, outfile)
```
